```python
import numpy as np
import pandas as pd
from annoy import AnnoyIndex
import csv
```

```python
train_data = pd.read_csv("./train.csv")
# data を np_array 変換
datas = train_data.values
# score のリストを保持
scores = train_data["Score"].values
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_data.shape
```

```python
train_data = train_data.drop("ID", axis=1)
train_data = train_data.drop("Score", axis=1)
```

```python
datas = train_data.values
```

```python
# annoy のインスタンス生成
annoy_model = AnnoyIndex(train_data.shape[1])
# インデックスの作成
for index, feature in enumerate(datas):
    annoy_model.add_item(index, feature)
```

```python
# モデルのビルド
annoy_model.build(100)
```

```python
# テストデータの読み取り
test_data = pd.read_csv("./test.csv")
```

```python
test_data.head()
```

```python
test_id = test_data["ID"].values
test_data = test_data.drop("ID", axis=1)
```

```python
test_datas = test_data.values
```

```python
result = list()
for f in test_datas:
    predict_index = annoy_model.get_nns_by_vector(f, 1, search_k=100)
    result.append(scores[predict_index])
```

```python
predict_csv = list()
for index, id in enumerate(test_id):
    tmp_list = list()
    tmp_list.append(id)
    tmp_list.append(result[index][0])
    predict_csv.append(tmp_list)
```

```python
# csv 出力
with open("predict.csv", "w", encoding="Shift_jis") as f:
    writer = csv.writer(f, lineterminator="\n")
    writer.writerows(predict_csv)
```
